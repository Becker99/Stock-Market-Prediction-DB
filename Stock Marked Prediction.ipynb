# Schlusskurs
Der Schlusskurs ist der letzte Kurs, zu dem eine Aktie während des regulären Handelstages gehandelt wird. Der Schlusskurs einer Aktie ist die Standard-Benchmark, die von den Anlegern verwendet wird, um die Entwicklung der Aktie im Laufe der Zeit zu verfolgen.

```python
# Let's see a historical view of the closing price
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    company['Adj Close'].plot()
    plt.ylabel('Adj Close')
    plt.xlabel(None)
    plt.title(f"Closing Price of {tech_list[i - 1]}")
    
plt.tight_layout()
```

# Volumen der Verkäufe
Das Volumen ist die Menge eines Vermögenswerts oder Wertpapiers, die über einen bestimmten Zeitraum, häufig im Laufe eines Tages, den Besitzer wechselt. Das Aktienhandelsvolumen bezieht sich beispielsweise auf die Anzahl der Aktien eines Wertpapiers, die zwischen dem täglichen Eröffnungs- und Schlusskurs gehandelt werden. Das Handelsvolumen und die Veränderungen des Volumens im Laufe der Zeit sind wichtige Inputs für technische Trader.

```python
# Now let's plot the total volume of stock being traded each day
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    company['Volume'].plot()
    plt.ylabel('Volume')
    plt.xlabel(None)
    plt.title(f"Sales Volume for {tech_list[i - 1]}")
    
plt.tight_layout()
```

# Wie hoch war der gleitende Durchschnitt der verschiedenen Aktien?
Der gleitende Durchschnitt (GDA) ist ein einfaches technisches Analyseinstrument, das die Preisdaten glättet, indem es einen ständig aktualisierten Durchschnittspreis bildet. Der Durchschnitt wird über einen bestimmten Zeitraum ermittelt, z. B. 10 Tage, 20 Minuten, 30 Wochen oder einen beliebigen Zeitraum nach Wahl des Händlers.

```python
ma_day = [10, 20, 50]

for ma in ma_day:
    for company in company_list:
        column_name = f"MA for {ma} days"
        company[column_name] = company['Adj Close'].rolling(ma).mean()
        

fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

DB[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[0,0])
axes[0,0].set_title('Deutsche Bank')

JPM[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[0,1])
axes[0,1].set_title('JP Morgan')

HSBC[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[1,0])
axes[1,0].set_title('HSBC')

WFC[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[1,1])
axes[1,1].set_title('Wells Fargo')

fig.tight_layout()
```

Aus dem Diagramm geht hervor, dass die besten Werte für die Messung des gleitenden Durchschnitts 10 und 20 Tage sind, da wir so die Trends in den Daten ohne Rauschen erfassen können.

# Wie hoch war die durchschnittliche tägliche Rendite der Aktie?
Nachdem wir nun einige grundlegende Analysen durchgeführt haben, wollen wir nun etwas tiefer einsteigen. Wir werden jetzt das Risiko der Aktie analysieren. Dazu müssen wir uns die täglichen Veränderungen der Aktie genauer ansehen, und nicht nur ihren absoluten Wert. Verwenden wir nun Pandas, um die täglichen Renditen der Apple-Aktie zu ermitteln.

```python
# We'll use pct_change to find the percent change for each day
for company in company_list:
    company['Daily Return'] = company['Adj Close'].pct_change()

# Then we'll plot the daily return percentage
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

DB['Daily Return'].plot(ax=axes[0,0], legend=True, linestyle='--', marker='o')
axes[0,0].set_title('Deutsche Bank')

JPM['Daily Return'].plot(ax=axes[0,1], legend=True, linestyle='--', marker='o')
axes[0,1].set_title('JP Morgan')

HSBC['Daily Return'].plot(ax=axes[1,0], legend=True, linestyle='--', marker='o')
axes[1,0].set_title('HSBC')

WFC['Daily Return'].plot(ax=axes[1,1], legend=True, linestyle='--', marker='o')
axes[1,1].set_title('Wells Fargo')

fig.tight_layout()
```

Großartig, jetzt wollen wir uns mit Hilfe eines Histogramms einen Gesamtüberblick über die durchschnittliche tägliche Rendite verschaffen. Wir werden seaborn verwenden, um sowohl ein Histogramm als auch ein KDE-Diagramm für dieselbe Abbildung zu erstellen.

```python
plt.figure(figsize=(12, 9))

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    company['Daily Return'].hist(bins=50)
    plt.xlabel('Daily Return')
    plt.ylabel('Counts')
    plt.title(f'{company_name[i - 1]}')
    
plt.tight_layout()
```

# Wie war die Korrelation zwischen den Schlusskursen verschiedener Aktien?
Die Korrelation ist eine Statistik, die das Ausmaß misst, in dem sich zwei Variablen im Verhältnis zueinander bewegen, wobei der Wert zwischen -1,0 und +1,0 liegen muss. Die Korrelation misst den Zusammenhang, sagt aber nichts darüber aus, ob x die Ursache für y ist oder umgekehrt - oder ob der Zusammenhang durch einen dritten Faktor verursacht wird[1].

Was wäre nun, wenn wir die Renditen aller Aktien in unserer Liste analysieren wollten? Machen wir weiter und erstellen einen DataFrame mit allen ['Close']-Spalten für jeden der Aktien-DataFrames.

```python
# Grab all the closing prices for the tech stock list into one DataFrame

closing_df = pdr.get_data_yahoo(tech_list, start=start, end=end)['Adj Close']

# Make a new tech returns DataFrame
tech_rets = closing_df.pct_change()
tech_rets.head()
```

Nun können wir die tägliche prozentuale Rendite von zwei Aktien vergleichen, um zu prüfen, inwieweit sie miteinander korrelieren. Betrachten wir zunächst eine Aktie im Vergleich zu sich selbst.

```python
# Comparing Google to itself should show a perfectly linear relationship
sns.jointplot(x='DB', y='DB', data=tech_rets, kind='scatter', color='seagreen')
```

```python
# We'll use joinplot to compare the daily returns of Allianz and Commerzbank
sns.jointplot(x='DB', y='JPM', data=tech_rets, kind='scatter')
plt.show()
```

Wir sehen also, dass, wenn zwei Aktien perfekt (und positiv) miteinander korreliert sind, eine lineare Beziehung zwischen ihren täglichen Renditewerten bestehen sollte.

Mit Seaborn und Pandas ist es sehr einfach, diese Vergleichsanalyse für jede mögliche Kombination von Aktien in unserer Tickerliste für Technologieaktien zu wiederholen. Wir können sns.pairplot() verwenden, um diese Darstellung automatisch zu erstellen

```python
# We can simply call pairplot on our DataFrame for an automatic visual analysis 
# of all the comparisons

sns.pairplot(tech_rets, kind='reg')
```

Oben sehen wir alle Beziehungen zwischen den täglichen Renditen aller Aktien. Ein kurzer Blick zeigt eine interessante Korrelation zwischen den Tagesrenditen von Google und Amazon. Es könnte interessant sein, diesen individuellen Zusammenhang zu untersuchen.

Während die Einfachheit des Aufrufs von sns.pairplot() fantastisch ist, können wir auch sns.PairGrid() verwenden, um die volle Kontrolle über die Abbildung zu haben, einschließlich der Frage, welche Art von Plots in die Diagonale, das obere Dreieck und das untere Dreieck passen. Im Folgenden finden Sie ein Beispiel für die Nutzung der vollen Leistungsfähigkeit von seaborn, um dieses Ergebnis zu erzielen.

```python
# Set up our figure by naming it returns_fig, call PairPLot on the DataFrame
return_fig = sns.PairGrid(tech_rets.dropna())

# Using map_upper we can specify what the upper triangle will look like.
return_fig.map_upper(plt.scatter, color='purple')

# We can also define the lower triangle in the figure, inclufing the plot type (kde) 
# or the color map (BluePurple)
return_fig.map_lower(sns.kdeplot, cmap='cool_d')

# Finally we'll define the diagonal as a series of histogram plots of the daily return
return_fig.map_diag(plt.hist, bins=30)
```

```python
# Set up our figure by naming it returns_fig, call PairPLot on the DataFrame
returns_fig = sns.PairGrid(closing_df)

# Using map_upper we can specify what the upper triangle will look like.
returns_fig.map_upper(plt.scatter,color='purple')

# We can also define the lower triangle in the figure, inclufing the plot type (kde) or the color map (BluePurple)
returns_fig.map_lower(sns.kdeplot,cmap='cool_d')

# Finally we'll define the diagonal as a series of histogram plots of the daily return
returns_fig.map_diag(plt.hist,bins=30)
```

Schließlich könnten wir auch ein Korrelationsdiagramm erstellen, um die tatsächlichen numerischen Werte für die Korrelation zwischen den täglichen Renditen der Aktien zu erhalten. Vergleicht man die Schlusskurse, so zeigt sich eine interessante Beziehung zwischen Deutsche Bank und JP Morgan.

```python
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
sns.heatmap(tech_rets.corr(), annot=True, cmap='summer')
plt.title('Correlation of stock return')

plt.subplot(2, 2, 2)
sns.heatmap(closing_df.corr(), annot=True, cmap='summer')
plt.title('Correlation of stock closing price')
```

Wie wir in unserem PairPlot vermutet haben, sehen wir hier numerisch und visuell, dass Microsoft und Amazon die stärkste Korrelation der täglichen Aktienrendite aufweisen. Es ist auch interessant zu sehen, dass alle Technologieunternehmen positiv korreliert sind.

# Wie viel Wert setzen wir mit einer Investition in eine bestimmte Aktie aufs Spiel?
Es gibt viele Möglichkeiten, das Risiko zu quantifizieren. Eine der grundlegendsten Methoden, bei der wir die Informationen über die täglichen prozentualen Renditen verwenden, ist der Vergleich der erwarteten Rendite mit der Standardabweichung der täglichen Renditen.

```python
rets = tech_rets.dropna()

area = np.pi * 20

plt.figure(figsize=(10, 8))
plt.scatter(rets.mean(), rets.std(), s=area)
plt.xlabel('Expected return')
plt.ylabel('Risk')

for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(label, xy=(x, y), xytext=(50, 50), textcoords='offset points', ha='right', va='bottom', 
                 arrowprops=dict(arrowstyle='-', color='blue', connectionstyle='arc3,rad=-0.3'))
```

# Vorhersage des Schlusskurses der Aktie der Deutschen Bank:

```python
# Get the stock quote
df = pdr.get_data_yahoo('DB', start='2012-01-01', end=datetime.now())
# Show teh data
df
```

```python
plt.figure(figsize=(16,6))
plt.title('Close Price History')
plt.plot(df['Close'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.show()
```

```python
# Create a new dataframe with only the 'Close column 
data = df.filter(['Close'])
# Convert the dataframe to a numpy array
dataset = data.values
# Get the number of rows to train the model on
training_data_len = int(np.ceil( len(dataset) * .95 ))

training_data_len
```

```python
# Scale the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)

scaled_data
```

```python
# Create the training data set 
# Create the scaled training data set
train_data = scaled_data[0:int(training_data_len), :]
# Split the data into x_train and y_train data sets
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
    if i<= 61:
        print(x_train)
        print(y_train)
        print()
        
# Convert the x_train and y_train to numpy arrays 
x_train, y_train = np.array(x_train), np.array(y_train)

# Reshape the data
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
# x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM

# Build the LSTM model
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(x_train, y_train, batch_size=1, epochs=1)
```

```python
# Create the testing data set
# Create a new array containing scaled values from index 1543 to 2002 
test_data = scaled_data[training_data_len - 60: , :]
# Create the data sets x_test and y_test
x_test = []
y_test = dataset[training_data_len:, :]
for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])
    
# Convert the data to a numpy array
x_test = np.array(x_test)

# Reshape the data
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1 ))

# Get the models predicted price values 
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

# Get the root mean squared error (RMSE)
rmse = np.sqrt(np.mean(((predictions - y_test) ** 2)))
rmse
```

```python
# Plot the data
train = data[:training_data_len]
valid = data[training_data_len:]
valid['Predictions'] = predictions
# Visualize the data
plt.figure(figsize=(16,6))
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.plot(train['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```

```python
# Letzte 60 Tage der Daten, einschließlich des letzten Testdatums
last_60_days = scaled_data[-60:]

# Erstellen einer neuen Sequenz mit den letzten 60 Tagen der Daten
x_input = np.array(last_60_days).reshape(1, -1)

# Umformen der Daten, um den Anforderungen des LSTM-Modells zu entsprechen
x_input = np.reshape(x_input, (1, x_input.shape[1], 1))

# Vorhersage für den nächsten Tag treffen
next_day_prediction = model.predict(x_input)

# Rücktransformation der Vorhersage in die ursprüngliche Skala
next_day_price = scaler.inverse_transform(next_day_prediction)

print("Vorhersagter Preis für die Aktie morgen:", next_day_price[0][0])
```
